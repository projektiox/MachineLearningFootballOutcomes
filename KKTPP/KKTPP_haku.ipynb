```python
import requests
import pandas as pd

# Step 1: Fetch the main list
url = "https://prodkuntarest.westeurope.cloudapp.azure.com/rest/v1/json/aineistot"
response = requests.get(url)
data = response.json()

aineistot = data["aineistot"]

# Step 2: Filter entries based on fields
filtered = [
    item for item in aineistot
    if item.get("hyvaksymisvaihe", "").lower() == "lopullinen"
    and item.get("raportointikokonaisuus", "") == "KKTPP"
    and item.get("raportointikausi", "") == "2023"
]

# Step 3: Fetch data from each 'tunnusluvut' URL
records = []
for item in filtered:
    url = item.get("tunnusluvut")
    try:
        resp = requests.get(url)
        details = resp.json()
        if isinstance(details, list):
            for entry in details:
                records.append({
                    "arvo": entry.get("arvo"),
                    "tunnusluku": entry.get("tunnusluku"),
                    "ytunnus": entry.get("ytunnus"),
                    "source_url": url
                })
    except Exception as e:
        print(f"Error fetching {url}: {e}")

# Step 4: Convert to DataFrame
df = pd.DataFrame(records)

# Display the result
print(df.head())
```

```python
import requests
import pandas as pd

# Step 1: Fetch the main list
url = "https://prodkuntarest.westeurope.cloudapp.azure.com/rest/v1/json/aineistot"
response = requests.get(url)
data = response.json()

aineistot = data["aineistot"]

# Step 2: Filter entries
filtered = [
    item for item in aineistot
    if item.get("hyvaksymisvaihe", "").lower() == "lopullinen"
    and item.get("raportointikokonaisuus", "") == "KKTPP"
    and item.get("raportointikausi", "") == "2023"
]

# Step 3: Skip problematic URL
excluded_url = "https://prodkuntarest.westeurope.cloudapp.azure.com/rest/v1/json/dokumentti/lopullinen/KKTPP/0927140-5/2023"

records = []
for item in filtered:
    url = item.get("tunnusluvut")
    if url == excluded_url:
        print(f"Skipping excluded URL: {url}")
        continue
    try:
        resp = requests.get(url)
        details = resp.json()
        if isinstance(details, list):
            for entry in details:
                records.append({
                    "arvo": entry.get("arvo"),
                    "tunnusluku": entry.get("tunnusluku"),
                    "ytunnus": entry.get("ytunnus"),
                    "source_url": url
                })
    except Exception as e:
        print(f"Error fetching {url}: {e}")

# Step 4: Convert to DataFrame
df = pd.DataFrame(records)

# Display the result
print(df.head())
```
